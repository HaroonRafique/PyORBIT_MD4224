```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import numpy as np
import scipy.io as sio 
import matplotlib.cm as cm
import PyNAFF as pnf
```

```python
plt.rcParams['figure.figsize'] = [8.0, 4.0]
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

plt.rcParams['font.size'] = 10
plt.rcParams['legend.fontsize'] = 'medium'
plt.rcParams['lines.linewidth'] = 1
```

```python
def add_input_file(dd, filename, label):
	f = filename
	p = dict()
	sio.loadmat(f, mdict=p)
	dd[label] = p	
	print '\tAdded output data from ', filename, '\t dictionary key: ', label
	return dd
```

```python
#Horizontal Scan No Space Charge

# Create dd dictionary
dd3 = dict()

Case = 'H'

#dd3 = add_input_file(dd3, '01_NoSC_26/output/output.mat', '6.26')
#dd3 = add_input_file(dd3, '01_NoSC_25/output/output.mat', '6.25')
dd3 = add_input_file(dd3, '02_NoSC_25/output/output.mat', 'NoSC')
dd3 = add_input_file(dd3, '02_SC_25/output/output.mat', 'SC')
sc = ''
legend_label = 'Case'

main_label = 'Zero_Dispersion'
main_label2 = main_label + '_zoom'
scaled_label = main_label + '_scaled'
turn_tot = 2200
zoom_turns = 15
turns = [0, 1, 10, 50, 100, 874, 2185, 2199]
betagamma = 2.492104532 * 0.9159915293879255
```

```python
dd3['SC'].keys()
```

```python
parameter = 'sig_x'

fig1 = plt.figure(facecolor='w', edgecolor='k')
ax1 = fig1.add_subplot(111)

tit = main_label + ' ' + sc + ' ' + parameter
ax1.set_title(tit);

ax1.set_ylabel(r'$\sigma_x$ [mm]');
ax1.set_xlabel('Turn [-]');

colors = cm.rainbow(np.linspace(0, 1, len(dd3.keys())))
c_it = int(0)

for key, value in sorted(dd3.iteritems()):
        ax1.plot(dd3[key]['turn'][0], dd3[key][parameter][0]*1000, label=key, color=colors[c_it]);
        c_it = c_it + 1;
        
ax1.legend(fontsize='x-small', loc=4, title=legend_label)
ax1.grid()
ax1.set_xlim(-1,2200)


savename = 'Plots/Param_' + parameter + '.png'
plt.savefig(savename, dpi = 800);
```
