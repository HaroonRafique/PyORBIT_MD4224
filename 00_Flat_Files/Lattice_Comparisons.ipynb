```python
# List of directories
dirs=[]

dirs.append('Original_Lattice/Flat_Files/H_07_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_08_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_09_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_10_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_11_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_12_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_13_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_14_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_15_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_16_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_17_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_18_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_19_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_20_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/H_21_24/original_leq_final_h.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_24/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_23/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_22/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_21/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_20/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_19/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_18/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_17/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_16/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_15/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_14/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_13/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_12/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_11/original_leq_final_v.tfs')
dirs.append('Original_Lattice/Flat_Files/V_21_10/original_leq_final_v.tfs')

# Create dict to hold data
OG = dict()

# Iterate over directories to create tree structure: H/V -> Tune
OG['H'] = dict()
OG['V'] = dict()

# Make dictionary structure
for ifile in dirs:
    plane = ifile.split('/')[2][0]
    if plane is 'V':
        tune = ifile.split('/')[2][5:]
        OG['V'][str(tune)] = dict()
        # print 'Plane ', plane, ' tune ', tune
    elif plane is 'H':
        tune = ifile.split('/')[2][2:4]
        OG['H'][str(tune)] = dict()
        # print 'Plane ', plane, ' tune ', tune

for keys in OG:
    for keys_again in OG[keys]:
        print keys, keys_again
    
# Iterate over list of directories
for infile in dirs:
    plane = infile.split('/')[2][0]
    if plane is 'V': tune = infile.split('/')[2][5:]
    elif plane is 'H': tune = infile.split('/')[2][2:4]
    # print 'start ', plane, tune

    # import data (cut header)
    fin = open(infile,'r').readlines()[90:]

    # Iterate over data
    sin = []
    betxin = []
    betyin = []
    dxin = []
    dyin = []
    xin = []
    pxin = []
    yin = []
    pyin = []

    for f in fin:
        sin.append(float(f.split()[1]))
        betxin.append(float(f.split()[2]))
        betyin.append(float(f.split()[3]))
        dxin.append(float(f.split()[4])*beta)
        dyin.append(float(f.split()[5])*beta)
        #dxin.append(float(f.split()[4]))
        #dyin.append(float(f.split()[5]))
        xin.append(float(f.split()[6]))
        pxin.append(float(f.split()[7]))
        yin.append(float(f.split()[8]))
        pyin.append(float(f.split()[9]))

    # Add data to dictionary
    # print 'end ', plane, tune
    
    OG[plane][tune]['s']=sin
    OG[plane][tune]['betx']=betxin
    OG[plane][tune]['bety']=betyin
    OG[plane][tune]['dx']=dxin
    OG[plane][tune]['dy']=dyin
    OG[plane][tune]['x']=xin
    OG[plane][tune]['px']=pxin
    OG[plane][tune]['y']=yin
    OG[plane][tune]['py']=pyin
```
