```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
import numpy as np
import scipy.io as sio 
import matplotlib.cm as cm
import imageio
#import PyNAFF as pnf
```

```python
plt.rcParams['figure.figsize'] = [4.0, 4.0]
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

plt.rcParams['font.size'] = 10
plt.rcParams['legend.fontsize'] = 'medium'
plt.rcParams['lines.linewidth'] = 1
```

```python
def add_input_file(dd, filename, label):
	f = filename
	p = dict()
	sio.loadmat(f, mdict=p)
	dd[label] = p
	print '\tAdded output data from ', filename, '\t dictionary key: ', label
	return dd
```

```python
# Load a file
dd = dict()
maxi = 51
#turns = np.linspace(1, maxi, maxi)
turns = range(1,maxi)
print turns
```

```python
for t in turns:
    if t <= 9:    inputs = 'mainbunch_00000' + str(int(t)) + '.mat'
    else:    inputs = 'mainbunch_0000' + str(int(t)) + '.mat'
    print inputs
    dd = add_input_file(dd, inputs, t)
```

```python
main_label = 'V_10'
sc = 'SbS'
legend_label = 'Turn'
```

```python
# access all particle x co-ordinates like this
dd[1]['particles']['x'][0][0][0]
```

```python
bw_gifnames_x_xp = []
hm_gifnames_x_xp = []
bw_gifnames_y_yp = []
hm_gifnames_y_yp = []
```

```python
parameter1 = 'x'
parameter2 = 'xp'
multi1 = 1E3
multi2 = 1E3

for key, value in sorted(dd.iteritems()):
    
    fig1, ax1 = plt.subplots(1, figsize=(6,6))

    tit = main_label + ' ' + sc + ' ' + parameter1 + ' ' + parameter2 + ' turn ' + str(key)
    
    ax1.set_title(tit);

    ax1.set_ylabel(r'x$^{\prime}$ [mrad]');
    ax1.set_xlabel('x [mm]');

    #colors = cm.rainbow(np.linspace(0, 1, len(dd.keys())))
    #c_it = int(0)

    s = np.ones(len(dd[key]['particles'][parameter1][0][0][0]))*0.1

    ax1.scatter(dd[key]['particles'][parameter1][0][0][0]*multi1, dd[key]['particles'][parameter2][0][0][0]*multi2, color='k', marker='.', s=s);
    #c_it = c_it + 1

    #ax1.legend(fontsize='x-small', loc=4, title=legend_label)
    ax1.grid(linewidth=0.1)
    ax1.set_xlim(-20, 20)
    ax1.set_ylim(-1, 1)

    savename = 'Plots/x_xp/' + parameter1 + ' ' +  parameter2 + 'turn_' + str(int(key)) + '_V_bw.png'
    
    bw_gifnames_x_xp.append(savename)
    
    fig1.savefig(savename, dpi = 100)
    plt.close()
```

```python
images = []
for filename in bw_gifnames_x_xp:
    images.append(imageio.imread(filename))
imageio.mimsave('Plots/x_xp/x_xp_bw.gif', images, duration=0.5)
```

```python
parameter1 = 'x'
parameter2 = 'xp'
multi1 = 1E3
multi2 = 1E3

for key, value in sorted(dd.iteritems()):
    
    fig1, ax1 = plt.subplots(1, figsize=(6,6))

    tit = main_label + ' ' + sc + ' ' + parameter1 + ' ' + parameter2 + ' turn ' + str(key)
    
    ax1.set_title(tit);

    ax1.set_ylabel(r'x$^{\prime}$ [mrad]');
    ax1.set_xlabel('x [mm]');

    x_min = -10
    x_max = 10
    y_min = -0.6
    y_max = 0.6
    bin_x = 128
    bin_y = 128

    s = np.ones(len(dd[key]['particles'][parameter1][0][0][0]))*0.1
    
    heatmap, xedges, yedges = np.histogram2d(dd[key]['particles'][parameter1][0][0][0]*multi1, dd[key]['particles'][parameter2][0][0][0]*multi2, bins=(bin_x, bin_y), range=[[x_min, x_max],[y_min, y_max]])
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
    
    ax1.imshow(heatmap, extent=extent, aspect=( (x_min - x_max)/(y_min - y_max) ) )

    #ax1.legend(fontsize='x-small', loc=4, title=legend_label)
    ax1.grid(linewidth=0.1)
    ax1.set_xlim(x_min, x_max)
    ax1.set_ylim(y_min, y_max)

    savename = 'Plots/x_xp/' +  parameter1 + ' ' +  parameter2 + 'turn_' + str(int(key)) + '_V_hm.png'
    hm_gifnames_x_xp.append(savename)
    
    fig1.savefig(savename, dpi = 100)
    plt.close()
```

```python
images = []
for filename in hm_gifnames_x_xp:
    images.append(imageio.imread(filename))
imageio.mimsave('Plots/x_xp/x_xp_hm.gif', images, duration=0.5)
```
